```python
#Dataframe van de benodigde stationsinformatie voor het maken van de visualisaties (met Plotly)
stations_data = {'code': ['ASD','UT','RTD','GVC','EHV','LEDN','AH','ASS','GV','AMF'], 
                 'stationType': ['MEGA_STATION', 'MEGA_STATION', 'MEGA_STATION', 'MEGA_STATION', 'MEGA_STATION', 
                                 'KNOOPPUNT_INTERCITY_STATION', 'KNOOPPUNT_INTERCITY_STATION', 'KNOOPPUNT_INTERCITY_STATION', 
                                 'KNOOPPUNT_INTERCITY_STATION', 'KNOOPPUNT_INTERCITY_STATION'], 
                 'lat': [52.378887,52.08889,51.924999,52.080276,51.443333,52.166111,51.985001,52.388889,52.069721,52.153889], 
                 'lng': [4.900278,5.110278,4.468889,4.325,5.481389,4.481667,5.899167,4.837778,4.3225,5.370555], 
                 'ingangsDatum': ['2017-02-01','2000-01-07','2017-02-01','2015-04-15','2019-12-15','2011-05-01','2016-06-12',
                                  '2015-03-18','2017-02-01','2019-12-15']}  
df_stations_data = pd.DataFrame(stations_data)
df_stations_data
```

```python
#Histogram van de verdeling van de type stations (gemaakt met Plotly)
fig = px.histogram(data_frame = df_stations_data, 
                   x = 'stationType', 
                   color = 'stationType', 
                   labels = {"stationType":"Type station"}, 
                   title = 'Histogram van de verdeling van de type stations')
fig.update_yaxes(title_text = 'Aantal stations')
fig.show()
```

```python
#Dataframe voor het maken van de tijdlijn met Plotly (let op de "finish datum" is gekozen als de presenteerdatum)
df_tijdlijn = pd.DataFrame([dict(Station="ADS", Start='2017-02-01', Finish='2021-10-01', Color = 'MEGA_STATION'),
                            dict(Station="UT", Start='2000-01-07', Finish='2021-10-01', Color = 'MEGA_STATION'),
                            dict(Station="RTD", Start='2017-02-01', Finish='2021-10-01', Color = 'MEGA_STATION'),
                            dict(Station="GVC", Start='2015-04-15', Finish='2021-10-01', Color = 'MEGA_STATION'),
                            dict(Station="EHV", Start='2019-12-15', Finish='2021-10-01', Color = 'MEGA_STATION'),
                            dict(Station="LEDN", Start='2011-05-01', Finish='2021-10-01', Color = 'KNOOPPUNT_INTERCITY_STATION'),
                            dict(Station="AH", Start='2016-06-12', Finish='2021-10-01', Color = 'KNOOPPUNT_INTERCITY_STATION'),
                            dict(Station="ASS", Start='2015-03-18', Finish='2021-10-01', Color = 'KNOOPPUNT_INTERCITY_STATION'),
                            dict(Station="GV", Start='2017-02-01', Finish='2021-10-01', Color = 'KNOOPPUNT_INTERCITY_STATION'),
                            dict(Station="AMF", Start='2019-12-15', Finish='2021-10-01', Color = 'KNOOPPUNT_INTERCITY_STATION')])
fig = px.timeline(df_tijdlijn, 
                  x_start = "Start", x_end = "Finish", y = "Station", 
                  color = "Color", 
                  labels = {"Color":"Type Station"}, 
                  title = 'Tijdlijn van de ingangsdata van de verschillende type stations')
fig.update_xaxes(title_text = 'Ingangsdatum')
fig.update_yaxes(autorange = "reversed") 
fig.show()
```
